## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare the data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

#3. Plot data
ax.plot(x, y)

#4. Customize Plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-aixs")

#5. Save and show
fig.savefig("images/simple-plot.png")
```

### Sublpots

```python
#subplots option 1
fig, ((ax0, ax1), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(20,10))
ax0.plot(x, x**2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_sales.keys(), nut_butter_sales.values())
ax4.hist(data);
```

```python
#subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(20,10))

ax[0,0].plot(x,x**2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_sales.keys(), nut_butter_sales.values())
ax[1,1].hist(data);
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,20), subplots=True);
```

```python
plt.style.use("seaborn-whitegrid")
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
#create figure and plot
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(over_50["age"],
                     over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

#customize the plot
ax.set(title="Heart Disease Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Targets")

#add the mean line
ax.axhline(over_50["chol"].mean(),
          linestyle='--',
          color="red");
```

```python
fig.savefig("images/heart-disease-cholesterol-practice.png")
```

## Heart Disease Analysis

```python
#create the figure and plot
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#ax0
#plot the data 
scatter = ax0.scatter(over_50["age"],
                      over_50["chol"],
                      c=over_50["target"],
                      cmap="summer")

#customize the plot
ax0.set(title="Heart Disease Cholesterol Level",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Targets")

#mean line
ax0.axhline(over_50["chol"].mean(),
            linestyle="--",
            color="red")

#ax1
#plot the data
scatter = ax1.scatter(over_50["age"],
                      over_50["thalach"],
                      c=over_50["target"],
                      cmap="summer")

#customize the plot
ax1.set(title="Heart Disease Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Targets")

#mean line
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--",
            color="red");

fig.suptitle("Heart Disease Analysis", fontsize="large", fontweight="bold");
```

```python
fig.savefig("images/heart-disease-analysis.png")
```
